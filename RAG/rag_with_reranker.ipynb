```python
from datasets import load_dataset
from langchain_community.docstore.document import Document

```

```python
question_answer = load_dataset("rag-datasets/rag-mini-bioasq", "question-answer-passages")
query_corpus = load_dataset("rag-datasets/rag-mini-bioasq", "text-corpus")
```

```python
import pandas as pd

df_q = pd.DataFrame(question_answer['test'])
df_c = pd.DataFrame(query_corpus['passages'])
```

```python
passage_data = df_c["passage"].to_list()
passage_id = df_c["id"].to_list()

corpus = []

for text, p_id in zip(df_c["passage"], df_c["id"]):
    doc = Document(
        page_content=text, 
        metadata={"id": p_id}
    )
    corpus.append(doc)
    
print(f"Sample Metadata: {corpus[0].metadata}")
print(f"Sampel pasasage : {corpus[0].page_content}")
print(len(corpus))

```
